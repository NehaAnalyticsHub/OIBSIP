```python
#Libraries used in below code

import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
df=pd.read_csv("data/CarPrice_Assignment.csv")
df

```

# Preprocessing

```python
df.isna().sum()

#no missing values 
```

```python
df.describe()
```

```python
df.info()
```

```python
#Converting categorical variables to numerical 
    
df=pd.get_dummies(df, columns=['fueltype', 'aspiration', 'doornumber', 'carbody', 'drivewheel', 'enginelocation',
                               'enginetype', 'cylindernumber', 'fuelsystem'])

#we have used one hot encoding for converting categroical columns to numerical
```

```python
#feature engineering 

df['horsepower_per_curbweight'] = df['horsepower'] / df['curbweight']
```

```python

df['car_volume'] = df['carlength'] * df['carwidth'] * df['carheight']
```

```python
df
```

```python
#normalising numerical features
# Normalizing numerical features is an important step in data preprocessing for machine learning. It involves scaling the features
# to a consistent range so that they have similar magnitudes.


features = ['wheelbase', 'carlength', 'carwidth', 'carheight', 'curbweight', 'enginesize', 'boreratio', 'stroke', 'compressionratio', 'horsepower', 'peakrpm', 'citympg', 'highwaympg', 'horsepower_per_curbweight', 'car_volume']
scaler = MinMaxScaler()
df[features] = scaler.fit_transform(df[features])
```

```python
#splitting data set into training and testing

#features
df1=df.drop(columns=['car_ID', 'CarName', 'price'])
#target
df2=df['price']

#80% training and 20%testing
X_train, X_test, y_train, y_test = train_test_split(df1, df2, test_size=0.2, random_state=42)

```

# Linear Regression

```python

model = LinearRegression()
```

```python

model.fit(X_train, y_train)
```

```python
#Evaluation

#prediction on test set
y_pred = model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = mse ** 0.5

#average absolute difference between the predicted values and the actual target values.
print(f"Mean Absolute Error: {mae:.2f}")
#average squared difference between the predicted values and the actual target values
print(f"Mean Squared Error: {mse:.2f}")
#on average, the predictions deviate from the actual values. 
print(f"Root Mean Squared Error: {rmse:.2f}")

#our result from linear regression model shows average result but if these values would be lower that it would be better result.
```

# Random Forest

```python

model = RandomForestRegressor(random_state=42)

    
```

```python

model.fit(X_train, y_train)
```

```python
#evaluation

y_pred = model.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = mse ** 0.5

print(f"Mean Absolute Error: {mae:.2f}")
print(f"Mean Squared Error: {mse:.2f}")
print(f"Root Mean Squared Error: {rmse:.2f}")

```

```python
#we can see random forest show better results as compare to linear regression model 

```
